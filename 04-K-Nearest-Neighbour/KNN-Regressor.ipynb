```python
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_features=3,n_samples=1000,random_state=67,noise=10)
```

```python
X,y
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=5,algorithm='auto')
```

```python
knn.fit(X_train,y_train)

y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error

print(f"accuracy : {r2_score(y_pred,y_test)}")
print(f"error : {mean_squared_error(y_pred,y_test)}")
print(f"error : {mean_absolute_error(y_pred,y_test)}")
```

```python
from sklearn.model_selection import GridSearchCV , RandomizedSearchCV

params = {
    'n_neighbors' : [3,4,5,6,7],
    'p' : [1,2],
    'algorithm' : ['auto','ball_tree','kd_tree']
}

grid = GridSearchCV(estimator=KNeighborsRegressor(),param_grid=params,cv=5,verbose=3,scoring='accuracy')
rgrid = RandomizedSearchCV(KNeighborsRegressor(),param_distributions=params,cv=5,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
rgrid.fit(X_train,y_train)
```

```python
print(grid.best_params_)
print(rgrid.best_params_)
```

```python
y_pred1 = grid.predict(X_test)
y_pred2 = rgrid.predict(X_test)

print(f"GridSearchCV : {r2_score(y_pred1,y_test)}")
print(f"RandomizedSearchCV : {r2_score(y_pred2,y_test)}")
```

```python
print(f"accuracy : {r2_score(y_pred1,y_test)}")
print(f"error : {mean_squared_error(y_pred1,y_test)}")
print(f"error : {mean_absolute_error(y_pred1,y_test)}")

print(f"accuracy : {r2_score(y_pred2,y_test)}")
print(f"error : {mean_squared_error(y_pred2,y_test)}")
print(f"error : {mean_absolute_error(y_pred2,y_test)}")
```
